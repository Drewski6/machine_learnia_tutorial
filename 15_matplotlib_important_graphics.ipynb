## Scatter Classification

- For our data, we are going to use a data set called "Iris flower data set".
    - {{https://en.wikipedia.org/wiki/Iris_flower_data_set}}
- This data set comes by default in the scikit-learn module for python. 
- Below we are importing the data from the sklearn database and assigning sections of the data to variables x, y, and names.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f"x contains {x.shape[0]} examples and {x.shape[1]} variables.")
print(f"There are {np.unique(y).size} classes.")
```

```python
print(x)
```

```python
print(y)
```

```python
print(names)
```

- With the following command we can make a scatter plot with 2 axis.
- The first axis (X axis) we say is x[:, 0] which is a single column slice of the x data set.
- The second axis (Y axis) we say is x[:, 1] which is a single column slice of the column with index 1 from the x data set.
- Last we set c (color) equal to y which represents the target values (in this case the species names)
    - y contains 3 different numbers (0, 1, and 2) so 3 colors are chosen to color the scatter plot.
- We can also add labels to our axis like in the previous lecture with the methods ylabel and xlabel.
- Lastly, to add some more visual representation to the graph we can pass 2 more keyword parameters to the scatter method:
    - alpha=0.5
        - Setting the alpha will allow us to change the tranparency of the points.
        - This is really nice visually, because it allows us to show density of our points.
    - s=x[:,2]*100
        - Setting s will change the size of the point.
        - We can set this equal to another point of our data set. In this case, column with index 2 indicates the petal size.
        - Lastly, we multiply by 100 for greater effect.

```python
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5 ,s=x[:, 2]*100)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

## 3D Graphs

- With this we are going to import a 3d projection to ax.
- When we do this, we have to use the Object Oriented Programming approach to drawing our plots.
- Adding the "%matplotlib" lets us render the data in the matplotlib widget which is an interactive 3 space where we can see the data in 3 dimensions.
    - In this widget, use right-mouse-button, left-mouse-button, and middle-mouse-button to navigate.

```python
# %matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

- Similar to the 2d plot, we can color our data points in a similar way by setting the keyword argument c equal to a color or variable.

```python
ax = plt.axes(projection="3d")
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
```

- We can also create graphs of complex mathmatic expressions.
- We start by defining a function "f" using a lambda function.

```python
f = lambda x, y: np.sin(x) + np.cos(x+y)
```

- Then we make some data using the linespace method.
- This creates a range of data with 100 points evenly spaced between 0 and 5. We do this for the X and Y axis.

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
```

- Then we are going to re-create X and Y by passing them through the meshgrid method.
- I don't totally understand what this does. Look it up if you need to know.
- We then pass that through the f function we just created to make our Z axis.

```python
X, Y = np.meshgrid(X, Y)

Z = f(X, Y)
```

```python
Z.shape
```

- Similar to before, we can make a plot and color it.

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
```

## Historgrams

- This is one of the more important graphs in machine learning and data science.
- This allows us to see the distribution of data that we're working with.
- With this, we can visualize if a distribution is normal or anormal or symetric or asymetric. 

- We can work with our data from earlier with the iris flower data.
- Below we can represent the first column of the x data (length of sepal).
    - X axis: We have the length (cm) of the sepal.
    - Y axis: We have the number of flowers that had this characteristic.

```python
plt.hist(x[:,0])
```

- With a few keyword parameters, we change how we represent the data:
    - bins: we can change the number of sections that we see.

```python
plt.hist(x[:,0], bins=30)
```

- We also have 2 dimensional histograms that show us the distribution of our data.
- Like before, we can edit the colors by changing "cmap"
- We can also add a colorbar with the colorbar method to add a key.

```python
plt.hist2d(x[:, 0], x[:, 1], cmap="Blues")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.colorbar()
```

- We can also use histograms to show us frequency of colors in images.
- Here is an image of a racoon imported from scipy.

```python
from scipy import misc
face = misc.face(gray=True)
plt.imshow(face, cmap="gray")
```

- We can take the data from this image and divide it into 255 bins (One for each shade of black, grey, or white)

```python
plt.hist(face.ravel(), bins=255)
plt.show()
```

## Contour Plot

- With a contour plot we can visualize 3 dimensional data easily to help us find local or global minimums or maximums.
- This is very useful in data science when it comes to model optimization.
- Many times, the problems AI/ML is really good at is optimizing under certain conditions. A contour plot helps us visualize this.
- Used to visualize cost function.

- Below we're using the data from our meshgrid.

```python
plt.contour(X, Y, Z)
```

- We can increase the number of lines by adding a 4th argument.
- We can also change the color with the "colors" keyword argument.
    - black gives us a contour who's positive values are solid lines and negative values are dotted lines.

```python
plt.contour(X, Y, Z, 30, colors="black")
```

- We also have a contourf method that allows us to draw our graphs with a filled in color gradient.

```python
plt.contourf(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```
